```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
```

```python
Job_name=[]
Company_name=[]
Place=[]
Date=[]
jtype=[]
experince=[]

for page_no in range(0,14):
    url=f'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20science&start={page_no}'
    page= requests.get(url)
    src=page.content
    soup= bs(src,'lxml')
    alljobs=soup.find_all('div',{"class":"css-1gatmva e1v1l3u10"})
 


    for i in alljobs:
        job_names= i.find("a",{"class":"css-o171kl"}).text.strip()
        Job_name.append(job_names)

        company_names=i.find("a",{"class":"css-17s97q8"}).text.strip()
        Company_name.append(company_names)

        place=i.find('span',{"class":"css-5wys0k"}).text.strip()
        Place.append(place)

        date=i.find('div',{"class":["css-4c4ojb",'css-do6t5g']}).text
        Date.append(date)

        Jtype=i.find('span',{"class":"css-1ve4b75 eoyjyou0"}).text.strip()
        jtype.append(Jtype)

        Jexp=i.find('div',{"class":"css-y4udm8"}).text
        experince.append(Jexp)
```

```python
data=[Job_name,Company_name,Place,Date,jtype,experince]
data=pd.DataFrame(data)
data=data.T
data.columns=["Job_name","Company_name","Place",'Date',"job_type","experience"]
data
```

```python
data.to_csv('wazzuf.csv',index=None)
```
